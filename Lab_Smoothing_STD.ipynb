```python
'''
(2) Median Filter
'''

def apply_median_filter(image,mask_size):
    new_image = np.array(image); 
    edgex = mask_size // 2
    edgey = mask_size // 2
    width,height = new_image.shape
    for x in range(edgex,width-edgex):
        for y in range(edgey,height-edgey):
            new_image[x,y] = np.median(new_image[x-edgex:x+edgex+1,y-edgey:y+edgey+1])
    return new_image


def dowork(path):
    images = []
    #read image as gray scaled
    original_img = io.imread(path) 
    gray_scaled =  rgb2gray(original_img)
    images.append(gray_scaled)
    #add noise
    noised_image = random_noise(gray_scaled,mode= 's&p',amount =0.05)
    images.append(noised_image)
    #apply built-in filter
    built_in_filter = median(noised_image) 
    images.append(built_in_filter)
    #apply our filter
    filtered_img = apply_median_filter(noised_image,3)
    images.append(filtered_img)
    show_images(images,['original','noised','built_in_filtered','our filter'])

PATHS = ['pout.tif',
    'tire.tif']


dowork(PATHS[0])
dowork(PATHS[1])
```
