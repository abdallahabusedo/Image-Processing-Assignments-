```python
from skimage.util import random_noise
import skimage.io as io
import skimage.color as imOps
import numpy as np
import matplotlib.pyplot as plt
import skimage.color as imcol
import skimage.filters as f

def show_images(images,titles=None):
    #This function is used to show image(s) with titles by sending an array of images and an array of associated titles.\n",
    # images[0] will be drawn with the title titles[0] if exists\n",
    # You aren't required to understand this function, use it as-is.\n",
    n_ims = len(images)
    if titles is None: titles = ['(%d)' % i for i in range(1,n_ims + 1)]
    fig = plt.figure()
    n = 1
    for image,title in zip(images,titles):
        a = fig.add_subplot(1,n_ims,n)
        if image.ndim == 2: 
            plt.gray()
        plt.imshow(image)
        a.set_title(title)
        n += 1
    fig.set_size_inches(np.array(fig.get_size_inches()) * n_ims)
    plt.show()
    
    
def Negative(image):
    img  = np.array(image)
    return 255 - img

def con_enh(image): # contrast enhancement
    img  = np.array(image)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i,j] < 101:
                print(img[i,j])
                img[i,j] = 50 + img[i,j] * 1.5
    return img      

def read_pic(path):
    original_img = io.imread(path) 
    return imcol.rgb2gray(original_img)
    
    
PATHS = ['imgs/Picture1.png',
        'imgs/Picture2.png']

# 1 
gray_scaled = read_pic(PATHS[0])
show_images([gray_scaled,Negative(gray_scaled)],['original','negative'])

# 2 
gray_scaled =io.imread(PATHS[1])
show_images([gray_scaled,con_enh(gray_scaled)],['original','contrast enhancemented'])



```
