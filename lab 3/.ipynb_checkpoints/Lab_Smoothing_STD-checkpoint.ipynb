```python
'''
(2) Median Filter
'''
def medianFilteringAlgorithim (img , filter_size):
    height = img.shape[0]
    width = img.shape[1]
    edgeX = filter_size // 2
    edgeY = filter_size // 2 
    data_final = np.array(img)
    for x in range(edgeX,(height - edgeX)):
        for y in range(edgeY,(width - edgeY)):
            data_final[x,y] = np.median(img[x-edgeX:x+edgeX+1,y-edgeY:y+edgeY+1]) 
    return data_final
```

```python

img = io.imread("pout.tif")
noiseimage = random_noise(img, mode='s&p', amount=0.05)
removed_noise = medianFilteringAlgorithim(noiseimage, 3) 
imgM = median(noiseimage)
show_images([img,noiseimage,removed_noise,imgM],["orignial image","noise image" , "final image" , "built in function"])
img = io.imread("tire.tif")
noiseimage = random_noise(img, mode='s&p', amount=0.05)
removed_noise = medianFilteringAlgorithim(noiseimage, 3) 
imgM = median(noiseimage)
show_images([img,noiseimage,removed_noise,imgM],["orignial image","noise image" , "final image" , "built in function"])
```
