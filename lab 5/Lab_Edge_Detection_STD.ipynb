```python
# 2. Custom Sobel (V-H-Both)
img = rgb2gray(io.imread('circuit.tif')) 

hx = np.array([
    [ -1,-2,-1],
    [  0, 0, 0],
    [  1, 2, 1]
]);


hy = np.array([
    [ -1, 0, 1],
    [ -2, 0, 2],
    [ -1, 0, 1]
]);

x = convolve2d(img, hx);
y = convolve2d(img, hy);


edge_strength = np.sqrt(np.square(x) + np.square(y));

threshold = 27;
edge_strength[edge_strength < threshold] = 0;

show_images([x,y,edge_strength],["Sobel_X", "Sobel_Y", "Sobel"]);

```

```python
# 3. Custom LoG
img = rgb2gray(io.imread('circuit.tif'))

thres=0 ;

f1 = np.array([
    [ 1, 1,1],
    [ 1,-8,1],
    [ 1, 1,1]
]);

f2 = np.array([
    [ 0, 1,0],
    [ 1,-4,1],
    [ 0, 1,0]
]);

ImgXF1 = convolve2d(img, f1);
ImgXF2 = convolve2d(img, f2);


ImgXF1[ImgXF1 < thres]= 0;
ImgXF2[ImgXF2 < thres]= 0;

show_images([ImgXF1,ImgXF2],["LoG1","LoG2"]);
```
