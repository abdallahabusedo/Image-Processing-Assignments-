```python
'''
1. Built-in Sobel, Prewill, Roberts
2. Custom Sobel (V-H-Both)
3. Custom LoG
'''
img = io.imread('circuit.tif')
img = rgb2gray(img)
```

```python
# 1. Built-in Sobel, Prewill, Roberts

edge_roberts = roberts(img)
edge_sobel  = sobel(img)
edge_prewitt = prewitt(img)

show_images([img, edge_sobel, edge_prewitt, edge_roberts],['original','sobel','prewitt', 'roberts'])
```

```python
# 2. Custom Sobel (V-H-Both)



edge_sobel_h = sobel_h(img)
edge_sobel_v = sobel_v(img)
show_images([img, edge_sobel_h,edge_sobel_v],['original','sobel_h','sobel_v'])

```

```python
# 3. Custom LoG
thres= 20

G_smooth = gaussian(img,1)
lap = laplace(G_smooth)
show_images([img, G_smooth, lap], ['original', 'G_smooth', 'laplace'])


```

```python
'''
Optional
1-Make edge detection
2-Multiply by factor < 1
3-Subtract (Orginal Image) - (Edge Detection Image)
'''
lap *= 0.5
new = img - lap
show_images([img, new], ['original','subtracted','esobel'])


```
